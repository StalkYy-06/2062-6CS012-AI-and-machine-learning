```python
def convert_units(conversion_type, value):
    """
    Converts a value to different measurement.

    Parameters:
    conversion_type (str): The type of conversion to perform.
                           Options:
                           - 'm_to_ft'
                           - 'ft_to_m'
                           - 'kg_to_lbs'
                           - 'lbs_to_kg'
                           - 'l_to_gal'
                           - 'gal_to_l'
    value (float): The numeric value to convert.

    Returns:
    float: Value converted.
    """

    if conversion_type == "m_to_ft":
        return value * 3.28084
    elif conversion_type == "ft_to_m":
        return value / 3.28084
    elif conversion_type == "kg_to_lbs":
        return value * 2.20462
    elif conversion_type == "lbs_to_kg":
        return value / 2.20462
    elif conversion_type == "l_to_gal":
        return value * 0.264172
    elif conversion_type == "gal_to_l":
        return value / 0.264172
    else:
        raise ValueError("Unsupported conversion type.")


def main():
    print("Unit Conversion Program")
    print("-----------------------")
    print("Choose conversion type:")
    print("1. Length")
    print("2. Weight")
    print("3. Volume")

    choice = input("Enter your choice (1/2/3): ")

    try:
        if choice == "1":
            print("\nLength Conversions:")
            print("1. Meters to Feet")
            print("2. Feet to Meters")
            sub_choice = input("Enter your choice (1/2): ")

            value = float(input("Enter the value to convert: "))

            if sub_choice == "1":
                result = convert_units("m_to_ft", value)
                print(f"{value} meters = {result:.4f} feet")
            elif sub_choice == "2":
                result = convert_units("ft_to_m", value)
                print(f"{value} feet = {result:.4f} meters")
            else:
                print("Invalid length conversion choice.")

        elif choice == "2":
            print("\nWeight Conversions:")
            print("1. Kilograms to Pounds")
            print("2. Pounds to Kilograms")
            sub_choice = input("Enter your choice (1/2): ")

            value = float(input("Enter the value to convert: "))

            if sub_choice == "1":
                result = convert_units("kg_to_lbs", value)
                print(f"{value} kg = {result:.4f} lbs")
            elif sub_choice == "2":
                result = convert_units("lbs_to_kg", value)
                print(f"{value} lbs = {result:.4f} kg")
            else:
                print("Invalid weight conversion choice.")

        elif choice == "3":
            print("\nVolume Conversions:")
            print("1. Liters to Gallons")
            print("2. Gallons to Liters")
            sub_choice = input("Enter your choice (1/2): ")

            value = float(input("Enter the value to convert: "))

            if sub_choice == "1":
                result = convert_units("l_to_gal", value)
                print(f"{value} liters = {result:.4f} gallons")
            elif sub_choice == "2":
                result = convert_units("gal_to_l", value)
                print(f"{value} gallons = {result:.4f} liters")
            else:
                print("Invalid volume conversion choice.")

        else:
            print("Invalid main choice. Please select 1, 2, or 3.")

    except ValueError:
        print("Error: Please enter a valid numeric value.")


if __name__ == "__main__":
    main()
```

```python
#task 2

def calculate_sum(numbers):
    """
    Calculates the sum of a list of numbers.

    Parameters:
    numbers (list of float): The list of numbers.

    Returns:
    float: The sum of the numbers.
    """
    return sum(numbers)


def calculate_average(numbers):
    """
    Calculates the average of a list of numbers.

    Parameters:
    numbers (list of float): The list of numbers.

    Returns:
    float: The average of the numbers.

    Raises:
    ValueError: If the list is empty.
    """
    if len(numbers) == 0:
        raise ValueError("Cannot calculate average of an empty list.")
    return sum(numbers) / len(numbers)


def find_maximum(numbers):
    """
    Finds the maximum value in a list of numbers.

    Parameters:
    numbers (list of float): The list of numbers.

    Returns:
    float: The maximum number.

    Raises:
    ValueError: If the list is empty.
    """
    if len(numbers) == 0:
        raise ValueError("Cannot find maximum of an empty list.")
    return max(numbers)


def find_minimum(numbers):
    """
    Finds the minimum value in a list of numbers.

    Parameters:
    numbers (list of float): The list of numbers.

    Returns:
    float: The minimum number.

    Raises:
    ValueError: If the list is empty.
    """
    if len(numbers) == 0:
        raise ValueError("Cannot find minimum of an empty list.")
    return min(numbers)


def main():
    print("Mathematical Operations on a List")
    print("---------------------------------")
    print("Choose an operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    choice = input("Enter your choice (1/2/3/4): ")

    try:
        user_input = input("Enter numbers separated by spaces: ").strip()

        if not user_input:
            raise ValueError("Input list cannot be empty.")

        numbers = [float(num) for num in user_input.split()]

        if choice == "1":
            result = calculate_sum(numbers)
            print(f"Sum = {result}")

        elif choice == "2":
            result = calculate_average(numbers)
            print(f"Average = {result}")

        elif choice == "3":
            result = find_maximum(numbers)
            print(f"Maximum = {result}")

        elif choice == "4":
            result = find_minimum(numbers)
            print(f"Minimum = {result}")

        else:
            print("Invalid operation choice.")

    except ValueError as e:
        print(f"Error: {e}")


if __name__ == "__main__":
    main()
```

```python
#task 2
def get_sublist(lst, start, end):
    """
    Returns a sublist from the given list between the specified start and end indices (inclusive).

    Parameters:
    lst (list): The original list.
    start (int): The starting index.
    end (int): The ending index (inclusive).

    Returns:
    list: A sublist from index start to end.

    Raises:
    ValueError: If indices are invalid.
    """
    if start < 0 or end >= len(lst):
        raise ValueError("Start or end index is out of range.")
    if start > end:
        raise ValueError("Start index cannot be greater than end index.")

    return lst[start:end + 1]


def main():
    try:
        user_input = input("Enter elements separated by spaces: ").strip()

        if not user_input:
            raise ValueError("Input list cannot be empty.")

        lst = user_input.split()

        start = int(input("Enter start index: "))
        end = int(input("Enter end index: "))

        result = get_sublist(lst, start, end)

        print("Sublist:", result)

    except ValueError as e:
        print("Error:", e)


if __name__ == "__main__":
    main()
```

```python
def remove_first_last(lst):
    """
    Removes the first and last elements of a list.

    Parameters:
    lst (list): The original list.

    Returns:
    list: A new list without the first and last elements.

    Raises:
    ValueError: If the list has fewer than 2 elements.
    """
    if len(lst) < 2:
        raise ValueError("List must contain at least two elements.")

    return lst[1:-1]


example = [1, 2, 3, 4, 5]
result = remove_first_last(example)
print(result)
```

### Task 5

```python
def get_first_n(lst, n):
    """
    Returns the first n elements from a list using slicing.

    Parameters:
    lst (list): The original list.
    n (int): Number of elements to extract.

    Returns:
    list: A new list containing the first n elements.

    Raises:
    ValueError: If n is negative.
    """
    if n < 0:
        raise ValueError("n cannot be negative.")

    return lst[:n]


example = [1, 2, 3, 4, 5]
result = get_first_n(example, 3)
print(result)
```

### Task 6

```python
def get_last_n(lst, n):
    """
    Returns the last n elements from a list using slicing.

    Parameters:
    lst (list): The original list.
    n (int): Number of elements to extract from the end.

    Returns:
    list: A new list containing the last n elements.

    Raises:
    ValueError: If n is negative.
    """
    if n < 0:
        raise ValueError("n cannot be negative.")

    return lst[-n:] if n != 0 else []


example = [1, 2, 3, 4, 5]
result = get_last_n(example, 2)
print(result)
```

# 4.3 Exercise on Nested List

### Task 1

```python
def flatten(lst):
    """
    Flattens a nested list into a single-dimensional list.

    Parameters:
    lst (list): The input nested list.

    Returns:
    list: A flattened list containing all elements in a single dimension.
    """
    flat_list = []
    for item in lst:
        if isinstance(item, list):
            flat_list.extend(flatten(item))
        else:
            flat_list.append(item)
    return flat_list


nested_list = [[1, 2], [3, 4], [5]]
result = flatten(nested_list)
print(result)
```

### Task 2

```python
def access_nested_element(lst, indices):
    """
    Accesses an element in a nested list given a list of indices.

    Parameters:
    lst (list): The nested list.
    indices (list of int): List of indices representing the path to the element.

    Returns:
    any: The element at the specified indices.

    Raises:
    IndexError: If any index is out of range.
    TypeError: If the accessed element is not a list when expecting more indices.
    """
    element = lst
    for idx in indices:
        if not isinstance(element, list):
            raise TypeError(f"Expected a list at index {idx}, but got {type(element).__name__}")
        element = element[idx]
    return element


nested_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
indices = [1, 2]
result = access_nested_element(nested_list, indices)
print(result)
```

### Task 3

```python
def sum_nested(lst):
    """
    Calculates the sum of all numbers in a nested list (any depth).

    Parameters:
    lst (list): The input nested list containing numbers or other nested lists.

    Returns:
    int/float: The sum of all elements in the nested list.
    """
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total

nested_list = [[1, 2], [3, [4, 5]], 6]
result = sum_nested(nested_list)
print(result)
```

### Task 4

```python
def remove_element(lst, elem):
    """
    Removes all occurrences of a specific element from a nested list.

    Parameters:
    lst (list): The input nested list.
    elem (any): The element to remove.

    Returns:
    list: A new nested list with all occurrences of elem removed.
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        elif item != elem:
            result.append(item)
    return result

nested_list = [[1, 2], [3, 2], [4, 5]]
elem_to_remove = 2
result = remove_element(nested_list, elem_to_remove)
print(result)
```

### Task 5

```python
def find_max(lst):
    """
    Finds the maximum element in a nested list (any depth).

    Parameters:
    lst (list): The input nested list containing numbers or other nested lists.

    Returns:
    int/float: The maximum element in the nested list.
    """
    max_val = float('-inf')

    for item in lst:
        if isinstance(item, list):
            max_val = max(max_val, find_max(item))
        else:
            max_val = max(max_val, item)

    return max_val


nested_list = [[1, 2], [3, [4, 5]], 6]
result = find_max(nested_list)
print(result)
```

### Task 6

```python
def count_occurrences(lst, elem):
    """
    Counts how many times a specific element appears in a nested list.

    Parameters:
    lst (list): The input nested list.
    elem (any): The element to count.

    Returns:
    int: The number of times elem appears in the nested list.
    """
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count


nested_list = [[1, 2], [2, 3], [2, 4]]
elem_to_count = 2
result = count_occurrences(nested_list, elem_to_count)
print(result)
```

### Task 7

```python
def deep_flatten(lst):
    """
    Flattens a deeply nested list into a single list.

    Parameters:
    lst (list): The input nested list of any depth.

    Returns:
    list: A new flattened list containing all elements in a single dimension.
    """
    flat_list = []
    for item in lst:
        if isinstance(item, list):
            flat_list.extend(deep_flatten(item))
        else:
            flat_list.append(item)
    return flat_list


nested_list = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
result = deep_flatten(nested_list)
print(result)
```

### Task 8

```python
def average_nested(lst):
    """
    Calculates the average of all elements in a nested list (any depth).

    Parameters:
    lst (list): The input nested list containing numbers or other nested lists.

    Returns:
    float: The average of all elements in the nested list.
    """
    def helper(sublist):
        """
        Returns the sum and count of elements in a nested list.
        """
        total = 0
        count = 0
        for item in sublist:
            if isinstance(item, list):
                sub_total, sub_count = helper(item)
                total += sub_total
                count += sub_count
            else:
                total += item
                count += 1
        return total, count

    total_sum, total_count = helper(lst)
    if total_count == 0:
        raise ValueError("Cannot calculate average of an empty list.")
    return total_sum / total_count


nested_list = [[1, 2], [3, 4], [5, 6]]
result = average_nested(nested_list)
print(result)
```

```python
# Matrix operation
A = np.array([[3, 4],
              [7, 8]])
B = np.array([[5, 3],
              [2, 1]])

A_inv = np.linalg.inv(A)
identity = np.dot(A, A_inv)
print("A * A^-1:\n", identity)

print("Is identity:", np.allclose(identity, np.eye(2)))

AB = np.dot(A, B)
BA = np.dot(B, A)
print("\nAB:\n", AB)
print("BA:\n", BA)
print("Are AB and BA equal?", np.array_equal(AB, BA))

AB_T = (np.dot(A, B)).T
B_TA_T = np.dot(B.T, A.T)
print("\n(AB)^T:\n", AB_T)
print("B^T * A^T:\n", B_TA_T)
print("Are they equal?", np.array_equal(AB_T, B_TA_T))
```
